# Audio as a numpy array

```python
import numpy as np
from scipy.io import wavfile
from matplotlib import pyplot as plt
import IPython.display as ipd

nu, y = wavfile.read('1.wav')   # nu = omega/2pi

y = np.mean(y, axis=1)  # Average channels into Mono
y = y[::4]              # Undersampling to reduce vector size
nu = nu//4              # Frequency should be adjusted accordingly
print("y shape: " + str(y.shape))
# data is now a vector indeed
plt.plot(np.arange(0,1,1/nu), y[:nu])  # Plot 1 second of audio data
plt.xlabel('time, seconds')
plt.title('waveform')

# Jupyter player
ipd.Audio(y.T, rate=nu)
```

# Image as a numpy array

```python
from matplotlib import pyplot as plt

X = plt.imread('Royal-Crescent.jpg')
print("X shape:" + str(X.shape))
X = X[::4, ::4, :]             # Downsample to reduce size
Xred = X * np.array([1,0,0])   # Red colour only
                               # Observation: numpy broadcasts arrays starting from the last dimension
plt.imshow(Xred)
plt.show()
Xgreen = X * np.array([0,1,0]) # Green only
plt.imshow(Xgreen)
plt.show()
Xblue  = X * np.array([0,0,1]) # Blue only
plt.imshow(Xblue)
plt.show()
```
