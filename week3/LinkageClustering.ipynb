```python
import numpy as np
from matplotlib import pyplot as plt

# Example from Figure 6 of lecture notes
X = np.array([[0  , 0],
              [4  ,-4],
              [4.4,-3.2],
              [6  ,-1.2],
              [6  ,-0.2]])

plt.plot(X[:,0], X[:,1], 'o', markersize=16)
for i in range(X.shape[0]):
    plt.annotate(i, X[i,:], fontsize=24)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
clust = linkage(X)
clust
dendrogram(clust)
plt.show()
kvec = fcluster(clust, t=3, criterion='maxclust')
print(kvec)
for k in range(1,4):
    plt.plot(X[kvec==k,0], X[kvec==k,1], 'o', markersize=16)
plt.show()
```

```python
from sklearn import datasets

# One of public benchmark datasets ("two moons")
X,C0 = datasets.make_moons(n_samples=20, noise=0.02)
plt.plot(X[C0==0,0], X[C0==0,1], 'o', X[C0==1,0], X[C0==1,1], 'o')
for i in range(X.shape[0]):
    plt.annotate(i, X[i,:], fontsize=18)
plt.title('True clustering')
plt.show()
```

```python
clust = linkage(X)
kvec = fcluster(clust, t=2, criterion='maxclust')
for k in range(1,3):
    plt.plot(X[kvec==k,0], X[kvec==k,1], 'o', markersize=16)
plt.show()
```
